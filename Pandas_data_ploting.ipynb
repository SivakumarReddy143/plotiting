```python
data={
    'number series':[1,2,3,4,5,6,7,8],
    'squares':[1,4,9,16,25,36,49,64]
}
```

```python
df1=pd.DataFrame(data)
```

```python
df1
```

```python
plt.plot(df1['number series'],df1['squares'],'--v')
plt.title('line plot')
plt.xlabel('numbers')
plt.ylabel('squares')
```

```python
plt.bar(df1['number series'],df1['squares'])
plt.title('bar chart')
plt.xlabel('numbers')
plt.ylabel('squares')
```

```python
plt.barh(df1['number series'],df1['squares'])
plt.title('Horizontal bar chart')
plt.ylabel('numbers')
plt.xlabel('squares')
```
